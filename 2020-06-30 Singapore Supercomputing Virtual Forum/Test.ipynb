```python
from qiskit.aqua.algorithms import Grover
from qiskit.aqua.components.oracles import TruthTableOracle 
```

```python
from qiskit.providers.basicaer import StatevectorSimulatorPy, QasmSimulatorPy
from qiskit.providers.aer import StatevectorSimulator
```

```python
oracle = TruthTableOracle('0100000000000000')
grover = Grover(oracle, num_iterations=2)
grover.set_backend(QasmSimulatorPy())
grover.run()
```

```python
grover.construct_circuit().draw('mpl')
```
